```python
import pandas as pd
import numpy as np
```

```python
cafe_clip = pd.read_csv('CAFE_CLIP.csv')
cafe_align = pd.read_csv('CAFE_ALIGN.csv')
fnd_clip = pd.read_csv('FND_CLIP.csv')
fnd_clip_retrain = pd.read_csv('FND_CLIP_Retrain.csv')
fnd_align = pd.read_csv('FND_ALIGN.csv')
fnd_align_retrain = pd.read_csv('FND_ALIGN_Retrain.csv')
matral = pd.read_csv('MATRAL.csv')
fcn_lp_clip = pd.read_csv('FCN_LP_CLIP.csv')
fcn_lp_align = pd.read_csv('FCN_LP_ALIGN.csv')
getral = pd.read_csv('GETRAL.csv')
# matral_ = pd.read_csv('MATRAL-.csv')
eann = pd.read_csv('EANN.csv')
```

```python
cafe_clip['Time'] = cafe_clip['Time'] + 1
cafe_align['Time'] = cafe_align['Time'] + 1
fnd_clip['Time'] = fnd_clip['Time'] + 1
fnd_clip_retrain['Time'] = fnd_clip_retrain['Time'] + 1
fnd_align['Time'] = fnd_align['Time'] + 1
fnd_align_retrain['Time'] = fnd_align_retrain['Time'] + 1
fcn_lp_clip['Time'] = fcn_lp_clip['Time'] + 1
fcn_lp_align['Time'] = fcn_lp_align['Time'] + 1
matral['Time'] = matral['Time'] + 1
getral['Time'] = getral['Time'] + 1
eann['Time'] = eann['Time'] + 1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设你已经有四个 DataFrame：cafe_clip, cafe_align, fnd_clip, matral

# 创建一个新的图形
plt.figure(figsize=(10, 6))

# 绘制每个 DataFrame 的折线图，并为每根折线指定颜色
# plt.plot(cafe_clip['Time'], cafe_clip['Accuracy'], label='CAFE+CLIP', marker='o', color='b')  # 蓝色
# plt.plot(cafe_align['Time'], cafe_align['Accuracy'], label='CAFE+ALIGN', marker='o', color='g')  # 绿色
plt.plot(fnd_clip['Time'], fnd_clip['Accuracy'], label='FND-CLIP', marker='o', color='r')  # 红色
# plt.plot(fnd_clip_retrain['Time'], fnd_clip_retrain['Accuracy'], label='FND-CLIP-Retrain', marker='o', color='c')  # 青色
plt.plot(fnd_align['Time'], fnd_align['Accuracy'], label='FND-ALIGN', marker='o', color='b')  # 蓝色
# plt.plot(fnd_align_retrain['Time'], fnd_align_retrain['Accuracy'], label='FND-ALIGN-Retrain', marker='o', color='y')  # 黄色
plt.plot(fcn_lp_clip['Time'], fcn_lp_clip['Accuracy'], label='FCN-LP-CLIP', marker='o', color='k')  # 黑色
plt.plot(fcn_lp_align['Time'], fcn_lp_align['Accuracy'], label='FCN-LP-ALIGN', marker='o', color='orange')  # 橙色
# plt.plot(getral['Time'], getral['Accuracy'], label='GETRAL', marker='o', color='#8A2BE2')  # 紫色（十六进制）
plt.plot(matral['Time'], matral['Accuracy'], label='MATRAL', marker='o', color='#8A2BE2')  # 灰色（RGB元组）
plt.plot(eann['Time'], eann['Accuracy'], label='EANN', marker='o', color='#FF6347')  # 番茄红（十六进制）

# 添加标题和标签
plt.title('Accuracy vs Time')
plt.xlabel('Time')
plt.ylabel('Accuracy')

# 设置x轴从1开始
plt.xlim(left=1)

# 如果需要，你也可以调整x轴的刻度
# 例如，设置刻度间隔为1
plt.xticks(range(1, int(cafe_clip['Time'].max()) + 1, 1))

# 添加图例
plt.legend()

# 显示网格
plt.grid()

# 保存图形
plt.savefig('early.jpg', dpi=300)

# 显示图形
plt.show()
```
